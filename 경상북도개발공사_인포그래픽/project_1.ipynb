```python
# 데이터 준비
sizes = place_info['수용세대']
labels = place_info['주택타입']
positions = [random.choice(range(1, 12)) for _ in range(11)]  # 데이터의 위치 지정
positions_y= [random.choice(range(1, 12)) for _ in range(11)]

# 버블 차트 그리기
plt.figure(figsize=(8, 6))
plt.scatter(positions, positions_y, s=sizes*100, alpha=0.5)  # y값을 모두 1로 설정하여 수평으로 그립니다.

# 각 버블 위에 문자 값을 표시
for i, label in enumerate(labels):
    plt.text(positions[i], positions_y[i], label, ha='center', va='center')

plt.axis('off')
# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 위에서 정의한 font_name 변수를 사용하여 한글 폰트 설정
plt.rc('font', family=font_name)

# Sample 데이터 생성
x = np.arange(len(place_info['주택타입']))  # the label locations
place_info_ = place_info[['주거전용면적','주거공용면적','지하주차장면적']]
width = 0.25  # the width of the bars
multiplier = 0

# 그래프 생성
fig, ax = plt.subplots()

# 그룹별로 바 차트 그리기
for attribute, measurement in place_info_.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('면적')
ax.set_title('주택 타입별 면적')
ax.set_xticks(x + width * (len(place_info_) - 1) / 2)
ax.set_xticklabels(place_info['주택타입'])
ax.legend(loc='upper left', ncol=3)
ax.set_ylim(0, 100)

plt.show()
```
